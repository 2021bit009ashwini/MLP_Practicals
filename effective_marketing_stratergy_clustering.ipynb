```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
dataset['CUST_ID'] = label_encoder.fit_transform(dataset['CUST_ID'])
dataset['BALANCE'] = label_encoder.fit_transform(dataset['BALANCE'])
dataset['BALANCE_FREQUENCY'] = label_encoder.fit_transform(dataset['BALANCE_FREQUENCY'])
dataset['PAYMENTS'] = label_encoder.fit_transform(dataset['PAYMENTS'])
dataset['MINIMUM_PAYMENTS'] = label_encoder.fit_transform(dataset['MINIMUM_PAYMENTS'])
dataset['PRC_FULL_PAYMENT'] = label_encoder.fit_transform(dataset['PRC_FULL_PAYMENT'])
dataset['TENURE'] = label_encoder.fit_transform(dataset['TENURE'])
dataset
```

```python
X = dataset.iloc[:, :10]
X
```

```python
#calculating the different K values
from sklearn.cluster import KMeans
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
#by Visualization
plt.plot(range(1, 11), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') #within cluster sum of squares
plt.show()
```

```python
k = 3  # Number of clusters
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(X)
# Get cluster labels assigned by the KMeans algorithm
labels = kmeans.labels_
```

```python
#Visualising the clusters
plt.figure(figsize=(8, 6))

# Extracting indices for data points belonging to cluster 0
cluster_0_indices = np.where(kmeans.labels_ == 0)[0]
plt.scatter(X.iloc[cluster_0_indices, 0], X.iloc[cluster_0_indices, 1], s=100, c='purple', label='Nike')

# Extracting indices for data points belonging to cluster 1
cluster_1_indices = np.where(kmeans.labels_ == 1)[0]
plt.scatter(X.iloc[cluster_1_indices, 0], X.iloc[cluster_1_indices, 1], s=100, c='orange', label='Addidas')

# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label='Centroids')

plt.title('Clustering Visualization')
plt.xlabel('Nike')
plt.ylabel('Addidas')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import silhouette_score
# Calculate silhouette score
silhouette_avg = silhouette_score(X, labels)
print("The average silhouette_score is :", silhouette_avg)
```

```python
plt.figure(figsize=(8, 6))
for cluster_id in set(labels):
    cluster_indices = np.where(labels == cluster_id)[0]
    plt.scatter(X.iloc[cluster_indices, 0], X.iloc[cluster_indices, 1], label=f'Cluster {cluster_id}')

plt.title('Clustering Visualization')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Extracting indices for data points belonging to each cluster
for cluster_id in set(kmeans.labels_):
    cluster_indices = np.where(kmeans.labels_ == cluster_id)[0]
    ax.scatter(X.iloc[cluster_indices, 0], X.iloc[cluster_indices, 1], X.iloc[cluster_indices, 2], label=f'Cluster {cluster_id}')

# Plotting the centroids of the clusters
ax.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], kmeans.cluster_centers_[:, 2], s=300, c='red', label='Centroids')

ax.set_title('Clustering Visualization (3D)')
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Feature 3')
ax.legend()

plt.show()
```

```python
cluster_labels = {}
for column_name in X.columns:
    column_data = X[column_name].values.reshape(-1, 1)  # Reshape to make it 2D
    kmeans.fit(column_data)
    cluster_labels[column_name] = kmeans.labels_
```

```python
def visualize_clusters(column_name, cluster_labels):
    plt.figure(figsize=(8, 6))
    unique_labels = np.unique(cluster_labels[column_name])
    colors = plt.cm.rainbow(np.linspace(0, 1, len(unique_labels)))

    for i, label in enumerate(unique_labels):
        cluster_data = X[column_name][cluster_labels[column_name] == label]
        plt.scatter(cluster_data.index, cluster_data.values, color=colors[i], label=f'Cluster {label}')

    plt.title(f'Clusters for {column_name}')
    plt.xlabel('Index')
    plt.ylabel(column_name)
    plt.legend()
    plt.show()

# Visualize clusters for each column
for column_name in X.columns:
    visualize_clusters(column_name, cluster_labels)
```

```python
import pandas as pd

# Sample customer_data DataFrame with 'customer_id', 'recency', 'frequency', and 'monetary_value' columns
customer_data = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 5],
    'recency': [10, 30, 5, 60, 20],
    'frequency': [5, 2, 10, 1, 3],
    'monetary_value': [100, 50, 200, 30, 80]
})

# Calculate quartiles for Recency, Frequency, and Monetary Value
quantiles = customer_data.quantile(q=[0.25, 0.5, 0.75])

# Function to assign RFM scores
def rfm_score(x, column, quantiles):
    if x <= quantiles[column][0.25]:
        return 4  # Highest score
    elif x <= quantiles[column][0.5]:
        return 3
    elif x <= quantiles[column][0.75]:
        return 2
    else:
        return 1  # Lowest score

# Assign RFM scores to each customer
customer_data['R'] = customer_data['recency'].apply(rfm_score, args=('recency', quantiles))
customer_data['F'] = customer_data['frequency'].apply(rfm_score, args=('frequency', quantiles))
customer_data['M'] = customer_data['monetary_value'].apply(rfm_score, args=('monetary_value', quantiles))

# Calculate RFM score by combining R, F, and M scores
customer_data['RFM Score'] = customer_data['R'] * 100 + customer_data['F'] * 10 + customer_data['M']

# Print the final RFM scores
print(customer_data[['customer_id', 'R', 'F', 'M', 'RFM Score']])
```

```python
def calculate_silhouette_score(n_clusters, dataset):
    # Instantiate the KMeans
    km = KMeans(n_clusters=n_clusters, random_state=42)
    
    # Fit the KMeans model and predict clusters
    labels = km.fit_predict(dataset)
    
    # Calculate Silhouette Score
    score = silhouette_score(dataset, labels, metric='euclidean')
    
    # Print the score
    print(f'Silhouette Average Score for {n_clusters} clusters: {score:.3f}')

# Try silhouette_score for k=5, 10, and 3 clusters
cluster_values = [2,3,5,10]

for n_clusters in cluster_values:
    calculate_silhouette_score(n_clusters, dataset)
```
