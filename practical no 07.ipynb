# The Elbow Method

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(numeric_data)
wcss=[]
distortions=[]
for i in range (1,21):
    kmeans=KMeans(n_clusters=i,init="k-means++",random_state=42)
    kmeans.fit(X_scaled)
    distortions.append(sum(np.min(cdist(X_scaled, kmeans.cluster_centers_,'euclidean'), axis=1)) / X.shape[0])
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(15,6))
plt.plot(range(1,21),wcss)
plt.xticks(np.arange(1,21,step=1))
plt.title("The Elbow Point Graph")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
from sklearn.preprocessing import OneHotEncoder
numeric_data = data.select_dtypes(include=['number'])

# If you have categorical columns that need encoding, do it here
# For example, if 'species' is a categorical column:
categorical_data = data.select_dtypes(exclude=['number'])
encoder = OneHotEncoder()
categorical_data_encoded = encoder.fit_transform(categorical_data)

# Combine numeric and encoded categorical data
X = np.concatenate((numeric_data.values, categorical_data_encoded.toarray()), axis=1)

# Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Now, you can proceed with KMeans clustering
km = KMeans(n_clusters=3)
y = km.fit_predict(X_scaled)
```

# Silhouette Method

```python
silscore=silhouette_score(X,y)
```

```python
print("The silhouette score is: ",silscore)
```
